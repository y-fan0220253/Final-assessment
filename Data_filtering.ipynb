```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Category': ['Paid Work', 'Unpaid household & care work'],
    'Men (mins/day)': [120, 60],
    'Women (mins/day)': [90, 120]
}

df = pd.DataFrame(data)
df.set_index('Category', inplace=True)

fig, ax = plt.subplots(figsize=(8, 6))
df.plot(kind='bar', ax=ax, color=['#4A90E2', '#E94A47'], width=0.4)

plt.title('The "Zero Value" Shift: ONS Time Use in the UK (2024)', fontsize=14, fontweight='bold')
plt.ylabel('Average Minutes per Day', fontsize=12)
plt.xticks(rotation=0, fontsize=11)
plt.legend(title='Gender', fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig('gender_data_gap.png', dpi=300)
plt.show()
```
